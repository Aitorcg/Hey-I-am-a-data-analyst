### Creating a model

```python
#That's the most important part of the project
#We create a model where we scored every player acording to his position and statistics
#We give a value to each variable we estimate is relevant for this position
#Finally, we add all the values and create a new column with the result

#That allows us to elect a player by some variables and not only one where he stands out
#We reduce the fault options
#The player we choice is a more thorough player
#we've been inspired by the movie Moneyball
```

```python
db.Position.value_counts()
```

```python
def FeatureEngineering(x):
    
    x['PercGoals']=x['Goals scored']*100/(x['Shots'] + 0.000001)
    
    return x
```

```python
def ScorePlayers(x):
    if(x.Position=='Goalkeeper'):
        x['Score'] = 0.1*x['Minutes played'] + 0.1*x['Games played'] -0.5*x['Goals conceded while player on pitch'] + 0.3*x['Recoveries']
    elif(x.Position=='Defender'):
        x['Score'] = 0.05*x['Minutes played'] + 0.05*x['Clearances'] -0.2*x['Own goals'] + 0.3*x['Recoveries'] + 0.2*x['Successful tackles'] - 0.2*x['Unssuccessful tackles']
    elif(x.Position=='Midfielder'):
        x['Score'] = 0.05*x['Minutes played'] + 0.05*x['Interceptions'] +0.2*x['Assists'] + 0.3*x['Long passes'] + 0.2*x['Short passes'] - 0.2*x['Duels lost']
    elif(x.Position=='Forward'):
        x['Score'] = 0.05*x['Minutes played'] + 0.05*x['Penalties won'] +0.2*x['Through balls'] + 0.3*x['Goals scored'] + 0.2*x['Shots on target'] - 0.2*x['Unsuccessful dribbles']

    return(x)
```

```python
db = db.apply(FeatureEngineering,axis=1)
db = db.apply(ScorePlayers,axis=1)
```
